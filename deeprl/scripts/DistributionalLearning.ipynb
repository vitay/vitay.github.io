```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", palette="muted", color_codes=True)
```

```python
f, axes = plt.subplots(1, 1, figsize=(7, 5), sharex=True)
sns.despine(left=True)
rs = np.random.RandomState(10)

# Generate a random univariate dataset
d = rs.normal(3.0, 2.0, size=1000)


sns.distplot(d, bins=np.linspace(-10, 10, 21), kde=False, norm_hist=True, 
             hist_kws={"linewidth": 2, "alpha": 0.7, "color": "b"})

plt.setp(axes, yticks=[], yticklabels=[], xticks=[-10, -5, 0, 5, 10], xlabel="Return")
plt.title("Value distribution")
plt.tight_layout()
plt.savefig("../img/distributionallearning.png")
plt.show()
```

```python
f, axes = plt.subplots(1, 1, figsize=(7, 5), sharex=True)
sns.despine(left=True)
rs = np.random.RandomState(10)

# Generate a random univariate dataset
d = rs.normal(3.0, 2.0, size=1000)
p = 1.2 + 0.9* d

sns.distplot(d, bins=np.linspace(-10, 10, 21), kde=False, norm_hist=True, label="$\mathcal{Z}(s, a)$",
             hist_kws={"linewidth": 2, "alpha": 0.7, "color": "b"})
sns.distplot(p, bins=0.2 + np.linspace(-10, 10, 21), kde=False, norm_hist=True, label="$\mathcal{T} \, \mathcal{Z}(s, a)$",
             hist_kws={"linewidth": 2, "alpha": 0.7, "color": "r"})

plt.setp(axes, yticks=[], yticklabels=[], xticks=[-10, -5, 0, 5, 10], xlabel="Return")
plt.legend()
plt.title("Value distribution")
plt.savefig("../img/distributionallearning2.png")
plt.tight_layout()
```
